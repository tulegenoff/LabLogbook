# TensorFlow Graphs

### Simple Example

There are two types of executions in Tensorflow. 
1. The first type is the eager execution, which is line by line. 
2. The second type is the graph execution, which creates a graph and then runs it. In Tensorflow2.0, this type of execution is implemented as AutoGraph through the Python decorator tf.function.


```python
n1 = tf.constant(1)
n2 = tf.constant(2)
```

```python
n3 = n1 + n2
```

```python
print(n3)
print(n3.numpy())
```

```python
@tf.function
def add_nodes(tensor1, tensor2):
    return tensor1 + tensor2

g_n3 = add_nodes(n1, n2)
print(g_n3)
print(g_n3.numpy())
```

###  Creating Tensors

Creating tensors with just tf.constant and tf.Variable can be tedious to create big tensors. Imagine you want to create random noise – well, you could do that by making a tensor with tf.zeros or tf.ones.

Just specify the shape in the format shape=[rows, columns] and a dtype, if it matters at all.


```python
tensor = tf.zeros(shape=[3, 4], dtype=tf.int32)
print(('Tensor full of zeros as int32, 3 rows and 4 columns:\n{0}').format(tensor.numpy()))
```

```python
tensor = tf.ones(shape=[5, 3], dtype=tf.float32)
print(('Tensor full of ones as float32, 5 rows and 3 columns:\n{0}').format(tensor.numpy()))
```

### Reshape Tensors

Reshape data with tf.reshape

One might have generated some random noise or have a dataset of images in different sizes, which needs to be one-dimensional in order to fit into some filter or convolution.

Use tf.reshape to reshape the images in whichever way we want. For e.g. define a tensor, and then reshape it into 8 columns with 1 row, instead of 2 columns with 4 rows.


```python
tensor = tf.constant([[3, 2],
                      [5, 2],
                      [9, 5],
                      [1, 3]])

reshaped_tensor = tf.reshape(tensor = tensor,
                             shape = [1, 8])

print(('Tensor BEFORE reshape:\n{0}').format(tensor.numpy()))

print(('\nTensor AFTER reshape:\n{0}').format(reshaped_tensor.numpy()))
```

### Casting Tensors

Cast tensors to other data types with tf.cast
Some functions in TensorFlow and Keras requires specific data types as inputs, and we can do that with tf.cast. Especially, casting from integer values to float values.

Make a tensor with the datatype of float32. Then cast this tensor to int, removing the comma and all decimals, while not rounding up or down.



```python
tensor = tf.constant([[3.1, 2.8],
                      [5.2, 2.3],
                      [9.7, 5.5],
                      [1.1, 3.4]], 
                      dtype=tf.float32)

tensor_as_int = tf.cast(tensor, tf.int32)

print(('Tensor with floats:\n{0}').format(tensor.numpy()))

print(('\nTensor cast from float to int (just remove the decimal, no rounding):\n{0}').format(tensor_as_int.numpy()))
```

## Linear Algebra Operations

Many algorithms or research needs linear algebra operations in order to implement and trying new things, e.g. making smaller changes in activation functions or optimizers. 

- Transpose tensor with tf.transpose
- Matrix Multiplication with tf.matmul
- Element-wise multiplication with tf.multiply
- Identity Matrix with tf.eye
- Determinant with tf.linalg.det
- Dot Product with tf.tensordot

### Transpose a tensor with tf.transpose



```python
A = tf.constant([[3,7],
                 [1, 9]])

A = tf.transpose(A)

print(('The transposed matrix A:\n{0}').format(A))
```

### Matrix multiplication with tf.matmul

```python
A = tf.constant([[3, 7],
                 [1, 9]])

v = tf.constant([[5],
                 [2]])

Av = tf.matmul(A, v)

print(('Matrix Multiplication of A and v results in a new Tensor:\n{0}').format(Av))
```

### Element-wise multiplication with tf.multiply

```python
Av = tf.multiply(A, v)

print(('Element-wise multiplication of A and v results in a new Tensor:\n{0}').format(Av))
```

### Make an identity matrix with tf.eye

```python
A = tf.constant([[3, 7],
                 [1, 9],
                 [2, 5]])

rows, columns = A.shape

print(('Get rows and columns in tensor A:\n{0} rows \n{1} columns').format(rows, columns))

A_identity = tf.eye(num_rows=rows,
                    num_columns=columns,
                    dtype=tf.int32)

print(('\nThe identity matrix of A:\n{0}').format(A_identity.numpy()))
```

### Find the determinant with tf.linalg.det

```python
A = tf.constant([[3, 7],
                 [1, 9]])

A = tf.dtypes.cast(A, tf.float32)

det_A = tf.linalg.det(A)

print(('The determinant of A:\n{0}').format(det_A))
```

### Find the dot product with tf.tensordot


```python
# Just run the next cell using 'Ctrl' + 'Enter'
```

```python
# Defining a 3x3 matrix
A = tf.constant([[32, 83, 5],
                 [17, 23, 10],
                 [75, 39, 52]])

# Defining another 3x3 matrix
B = tf.constant([[28, 57, 20],
                 [91, 10, 95],
                 [37, 13, 45]])
```

```python
dot_AB = tf.tensordot(a=A, b=B, axes=1).numpy()

print(('Dot product of A.B^T results in a new Tensor:\n{0}').format(dot_AB))

AB = tf.matmul(A, B)

print(('\nMatrix Multiplication of A.B^T results in a new Tensor:\n{0}').format(AB))
```
